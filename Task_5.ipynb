```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/preet/Desktop/Numpy and Pandas/titanic.csv")
df
```

```python
print(df.describe())
```

```python
print(df.info())
```

```python
print(df['Survived'].value_counts())
```

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']].dropna(), hue='Survived')
plt.show()
```

```python
# Histogram for age
df['Age'].hist(bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare Distribution by Passenger Class')
plt.show()
```

```python
#Scatterplot Age vs Fare colored by Survived
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Scatterplot of Age vs Fare by Survival')
plt.show()
```

```python
#Passenger class (Pclass) strongly influences fare paid and chances of survival
#Age is an important factor; children and younger passengers had higher survival rates
#Higher Fare correlates with increased survival probability, likely reflecting socio-economic status
#Survival rate is higher for females and 1st class passengers (implied by fare and class correlation
```

```python
#Summary:
#The Titanic dataset reveals clear social and demographic factors affecting survival. Passengers in 1st class who paid higher fares, and younger passengers had better survival odds, reflecting priority boarding and evacuation practices at the time. The strong correlations between Pclass, Fare, Age, and Survival highlight socioeconomic status and age as crucial determinants of survival on the Titanic.
```
